```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
titanic=['Sex','Cabin','Embarked']
data[titanic]=data[titanic].apply(le.fit_transform)
data
```

```python
data.drop(columns=['PassengerId','Name','Ticket'],axis=1,inplace=True)
data.columns
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(data.iloc[:,3:4])
data.iloc[:,3:4]=imputer.transform(data.iloc[:,3:4])
data
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,1:]
y=data.iloc[:,0:1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifer = RandomForestClassifier(random_state=42)
```

```python
classifer.fit(x_train, y_train)
```

```python
y_pred = classifer.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
